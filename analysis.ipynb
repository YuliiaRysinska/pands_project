```python
#using necessary libraries
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Loading iris dataset into panda data frame
iris = pd.read_csv('iris.csv')
```

```python
#view of general informations of data (first 10 rows)
print(iris.head(10))
```

```python
# information of attributes in the dataset
print(iris.info())
```

```python
# Basic statistical features of the dataset as a whole
print(iris.describe()) 
```

```python
#visual inspection of whether there is a relationship between independent and dependent metrics (categories)
sns.pairplot(data = iris, hue = 'species')
plt.savefig("pairplot.png")
```

```python
print(iris['species'].value_counts())
```

```python
# creating individual feature in Seaborn through a boxplot
sns.boxplot(x='species', y='petal_length', data=iris)
plt.savefig("species_boxplot.png")
```

```python
# Histogram of all measurements across all species.
iris.hist(figsize = (8, 6))
plt.savefig("species_histogram.png")
```

```python
# Correlations between measurements.
iris[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']].corr()
```

```python
# Visualising correlations with a heatmap.
sns.heatmap(iris[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']].corr(), annot = True, cmap = 'RdYlGn_r')
plt.savefig("corr_heatmap.png")
```

```python
# Closer look at scatter plots for petal measurements.
sns.FacetGrid(iris, hue = 'species', height=5).map(plt.scatter, 'petal_length', 'petal_width').add_legend()
plt.savefig("petal.png")
```

```python
# Closer look at scatter plots for sepal measurements.
sns.FacetGrid(iris, hue = 'species', height=5).map(plt.scatter, 'sepal_length', 'sepal_width').add_legend()
plt.savefig("sepal.png")
```
